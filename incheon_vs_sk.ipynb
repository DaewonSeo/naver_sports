```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('kbl_game_data_2015_2016.csv', delimiter=',')
```

```python
df
```

```python
df = df.replace("\t", "").replace("\n", "")
```

```python
team_jeon = df.iloc[:, 3:][df['team'] == '전자랜드']
```

```python
team_jeon.shape
```

```python
team_jeon
```

```python
team_jeon_x = team_jeon.iloc[:, 1:]
```

```python
team_jeon_y = team_jeon.iloc[:, 0]
```

```python
team_jeon_y
```

```python
from sklearn.cross_validation import train_test_split
```

```python
train_X , target_x, train_y, target_y = train_test_split(team_jeon_x, team_jeon_y, test_size=0.33, random_state=42)
```

```python
model1 = LinearRegression()
```

```python
model1.fit(train_X, train_y)
```

```python
model1.predict(target_x)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(model1.predict(target_x), target_y)
```
