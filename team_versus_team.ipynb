```python
import pandas as pd
import numpy as np
import csv
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('kbl_game_data_2015_2016.csv', delimiter=',')
```

```python
df = df.replace('\n', "").replace('\t', "")
```

```python
df['team'] = df['team'].replace('오리온', '오리온스')
```

```python
team_orions = df.iloc[:, 3:-1][df['team'] == '오리온스']
```

```python
team_lg = df.iloc[:, 3:][df['team'] == 'LG']
```

```python
model1 = LinearRegression()
```

```python
from sklearn.cross_validation import train_test_split
```

```python
team_orions
```

```python
team_orions_x = team_orions.iloc[:, 1:]
```

```python
team_orions_y = team_orions.iloc[:, 0]
```

```python
train_x, target_x, train_y, target_y = train_test_split(team_orions_x, team_orions_y, test_size=0.33, random_state=42)
```

```python
model1.fit(train_x, train_y)
```

```python
target_predict = model1.predict(target_x)
```

```python
target_predict
```

```python
target_y
```

```python
train_x
```

```python
model1.coef_
```

```python
model1.score(target_x, target_y)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(target_y,target_predict)
```
